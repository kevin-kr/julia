## Plotting
### Basics

```julia
using Pkg
Pkg.add("Plots")
using Plots
```

```julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numberofpirates = [45000, 20000, 15000, 5000, 400, 17];
```

```julia
gr()
```

```julia
plot(numberofpirates, globaltemperatures, label="line")
scatter!(numberofpirates, globaltemperatures, label="points") # mutating function
```

```julia
xlabel!("Number of Pirates [Approximate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
```
